```python
from fbprophet import Prophet
import pandas as pd 
import numpy as np
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python
# load data sets
houston = pd.read_csv(r'C:\Users\rafav\Documents\Flatiron\Module 4\mod 4 final project\houston_prophet.csv')
midland = pd.read_csv(r'C:\Users\rafav\Documents\Flatiron\Module 4\mod 4 final project\midland_prophet.csv')
bakersfield = pd.read_csv(r'C:\Users\rafav\Documents\Flatiron\Module 4\mod 4 final project\bakersfield_prophet.csv')
bakersfield_original = pd.read_csv(r'C:\Users\rafav\Documents\Flatiron\Module 4\mod 4 final project\Bakersfield.csv')
midland_original = pd.read_csv(r'C:\Users\rafav\Documents\Flatiron\Module 4\mod 4 final project\Midland.csv')
houston_original= pd.read_csv(r'C:\Users\rafav\Documents\Flatiron\Module 4\mod 4 final project\Houston.csv')
```

```python
houston
```

```python
houston.drop(['Unnamed: 0'], axis=1,inplace=True)
```

```python
model = Prophet() #instantiate Prophet
model.fit(houston); #fit the model with your dataframe
```

```python
future_houston = model.make_future_dataframe(periods=12, freq = 'm')
```

```python
forecast_data = model.predict(future_houston)
```

```python
forecast_data[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
model.plot(forecast_data)
```

```python
forecast_data.head()
```

```python
metric_houston= forecast_data.set_index('ds')[['yhat']].join(houston.set_index('ds').y).reset_index()
```

```python
metric_houston
```

```python
metric_houston.dropna(inplace=True)
```

```python
metric_houston.tail()
```

```python
r2_score(metric_houston.y, metric_houston.yhat)
```

```python
mean_squared_error(metric_houston.y, metric_houston.yhat)
```

```python
mean_absolute_error(metric_houston.y, metric_houston.yhat)
```
